```python
# Imports.
import pandas as pd
import csv as csv
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Import the datasets
edsvc_df = pd.read_csv(r"C:\Users\mitch\Desktop\Analytical BootCamp\Project-1\CSV_files\2017_Educational services_ecnbasic.csv")
edsvc_df.head()
```

```python
edsvc_slope, edsvc_int, edsvc_r, edsvc_p, edsvc_std_err = stats.linregress( edsvc_df['Number Employed'] , edsvc_df['Broadband (%)'] )
edsvc_fit = edsvc_slope * edsvc_df['Number Employed'] + edsvc_int
```

```python
plt.scatter( edsvc_df['Number Employed'] , edsvc_df['Broadband (%)'] , marker="o", facecolors="blue", edgecolors="red")
plt.title("Number Employed vs. Broadband %")
plt.xlabel("Number Employed")
plt.ylabel("Broadband (%)")
plt.plot( edsvc_df['Number Employed'] , edsvc_fit,"--")
plt.show()
print(f"The slope is {round(edsvc_slope,3)}.")
print(f"The intercept is {round(edsvc_int,3)}.")
print(f"The r-squared is {round(edsvc_r**2,3)}.")
```
